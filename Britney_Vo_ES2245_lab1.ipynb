### Part 3: Loops and Math (20 mins)

**Scenario:** You need to calculate the average magnitude of the earthquakes from Part 2. While Python has libraries for this, it is important to know how to do it manually first.

**Task:**
1. Initialize a variable `total_magnitude` to `0`.
2. Write a `for` loop that iterates through your `magnitudes` list.
3. Inside the loop, add each magnitude to `total_magnitude`.
4. After the loop, calculate the mean (average) by dividing the total by the number of earthquakes.
5. Print the average.

```python
# Part 3: Your Code Here

total_magnitude = 0

# Write your for loop here:

for value in magnitudes:
  total_magnitude += value

# Calculate mean

num_earthquakes = len(magnitudes)
average_magnitude = total_magnitude / num_earthquakes

# Print result

print(f"The average magnitude is: {average_magnitude}")
```

### Part 4: Conditional Logic (15 mins)

**Scenario:** You need to classify the earthquakes into categories to alert the public.
* **Magnitude < 3.0:** "Micro" (Not usually felt)
* **Magnitude >= 3.0:** "Minor" (Often felt)

**Task:**
1. Write a loop that goes through the `magnitudes` list.
2. Inside the loop, use an `if/else` statement to check the magnitude.
3. Print the magnitude and its classification (e.g., *"Magnitude 4.5 is Minor"*).

```python
# Part 4: Your Code Here

for mag in magnitudes:
    if mag < 3.0:
        print(f"Earthquake magnitude {mag} is micro. You may not notice it!")
    else:
        print(f"Earthquake magnitude {mag} is minor. Protect yourself!")
```

### Part 5: Putting it all together




**Scenario:** Geothermal gradient calculation.
The temperature at the surface is 15°C. The temperature increases by 25°C for every kilometer of depth (Geothermal Gradient).

**Task:**
1. Create a list of depths in km: `[0, 1, 2, 3, 4, 5]`.
2. Use a loop to calculate the temperature at each depth.
3. Store these temperatures in a new list called `temperatures`.
4. Print the final list.

```python
# Part 5: Your Code Here

depths = [0, 1, 2, 3, 4, 5]
temperatures = []

surface_temp = 15
gradient = 25

# Loop through depths, calculate temp, and append to temperatures list

for d in depths:
    temp = surface_temp + (gradient * d)
    temperatures.append(temp)

print(temperatures)
```
